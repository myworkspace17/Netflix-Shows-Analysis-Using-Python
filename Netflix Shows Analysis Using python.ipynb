# 2 In which year the highest number of shows is released . show them using bar chart

# dtypes

```python
df.dtypes
```

```python
df.head
```

```python
# Group by Release Year and Type, then count the occurrences
yearwise_type_count = df.groupby(['release_year', 'type']).size().unstack(fill_value=0)

# Display the grouped data
print(yearwise_type_count)
```

```python
# Plotting the year-wise type distribution (stacked bar plot)
yearwise_type_count.plot(kind='bar', stacked=True, figsize=(10, 6))

# Adding labels and title
plt.title('Year-wise Releases by Type (TV Show vs Movie)')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.legend(title='Type')


# Display the plot
plt.tight_layout()
plt.show()
```
